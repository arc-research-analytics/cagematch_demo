```python
import pandas as pd
import geopandas as gpd
```

```python
custom_colors_hex = [
    '#eff3ff',  # lightest shade
    '#bdd7e7',
    '#6baed6',
    '#3182bd',
    '#08519c'  # darkest shade
]

gdf = gpd.read_file('Data/Fulton_county.geojson')

# Create a color mapping using quantiles
gdf['density_hex'] = pd.qcut(gdf['rPopDensity_e21'], q=len(
    custom_colors_hex), labels=custom_colors_hex)

gdf
```
